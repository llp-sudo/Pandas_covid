```python
import pandas as pd
```

```python
df = pd.read_csv('../time_series_covid19_confirmed_global.csv')
```

```python
df.head()
```

```python
df_covid = df
```

```python
list(df)
```

```python
##Retirando colunas
```

```python
del df_covid['Long']
del df_covid['Lat']
```

```python
##Renomendo colunas
```

```python
df_covid = df_covid.rename(columns={'Province/State': 'Province', 'Country/Region':'Country'})
```

```python
df_covid
```

```python
##Transformando as criando a coluna Date a partir das colunas com datas
```

```python
df_covid = df_covid.melt(id_vars=["Province", "Country"], 
        var_name="Date")
```

```python
df_covid = df_covid.rename(columns={'value': 'cases'})
```

```python
df_covid
```

```python
##Criando DF contendo registros que tem provincias
```

```python
df_province = df_covid[df_covid['Province'].notnull()]
```

```python
df_province
```

```python
##Criando lista dos paÃ­ses
```

```python
lts_countries = df_province['Country'].unique()
```

```python
##Criando lista dos dias
```

```python
lts_dates = df_province['Date'].unique()
```

```python
##Juntado os valores da mesma data e mesmo paÃ­s - (retiramos a provincia dos paÃ­s e juntamos tudo em paÃ­s)
```

```python
total_case = df_province[(df_province['Country'] == 'Australia') & (df_province['Date'] == '1/28/20')].groupby(['Date']).sum()
```

```python
total_case['cases'][0]
```

```python
##Juntado os valores da mesma data e mesmo paÃ­s - (retiramos a provincia dos paÃ­s e juntamos tudo em paÃ­s)
## criando lista contendo os valores do merge dos paÃ­ses
```

```python
lts_tmp  = []
for c in lts_countries:
    print('=======================Juntando paÃ­ses e retirando a provincia ====================')
    print(f'Pais = {c}')
    for d in lts_dates:
        total_case = df_province[(df_province['Country'] == c) & (df_province['Date'] == d )].groupby(['Date']).sum()
        total_case = total_case['cases'][0]
        lts_tmp.append([c,d,total_case])
```

```python
lts_tmp
```

```python
##Criando DF a partir da lista
```

```python
df_tmp = pd.DataFrame(lts_tmp, columns=["Country", "Date", "cases"])
```

```python
#Vamos pegar o DF original, retirar as linhas que a Province != de NaN
```

```python
df2 = df_covid[df_covid['Province'].isnull()]
```

```python
del df2['Province']
```

```python
df2 = df2.loc[:, ~df2.columns.str.contains('^Unnamed')]
```

```python
frames = [df_tmp,df2 ]
```

```python
df_covid_final = pd.concat(frames)
```

```python
df_covid_final
```

```python
df_covid_final.to_csv('covid.csv', sep=',', header=True)
```
